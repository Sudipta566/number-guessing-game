```python
import random

def play_number_guessing_game():
    print("Welcome to the Guess the Number Challenge!")
    print("I have picked a number between 1 and 100.")
    print("Can you figure out what it is?")

    secret_number = random.randint(1, 100)

    total_attempts = 10

    for attempt in range(1, total_attempts + 1):
        try:
            user_guess = int(input(f"Attempt {attempt} of {total_attempts}: What's your guess? "))

            if user_guess == secret_number:
                print(f"Amazing! You guessed the number {secret_number} in {attempt} tries!")
                break
            elif user_guess < secret_number:
                print("Your guess is too low. Try again.")
            else:
                print("Your guess is too high. Try again.")
        except ValueError:
            print("Please enter a valid whole number.")
    else:
        print(f"Out of tries! The number I was thinking of was {secret_number}.")

if __name__ == "__main__":
    play_number_guessing_game()
```
